<h5>Useful functions</h5>

As several distributions will be explored, it will be useful to write a summarization function that will display the mean, variance and a histogram of the generated samples:

```python
number_of_samples = 100000
bins = 100

def graph_distribution(distribution, dist_name=''):
    """
    Output a histogram of the distribution
    """
    _ = plt.hist(distribution, 
             bins=bins)
    plt.xlabel('Sample Value')
    plt.ylabel('Count of Value')
    plt.title('Histogram of {} Distribution'.format(dist_name))
    plt.show()
    
def calculate_statistics(distribution):
    """
    Prints the mean and variance of the given distribution
    """
    mean = distribution.mean()
    var = distribution.var()
    
    print('Sample Mean: {}'.format(mean))
    print('Sample Variance: {}'.format(var))
    
def summarize(distribution, dist_name=''):
    """
    Prints the mean and variance of a distribution then 
    displays a histogram of the samples taken from the 
    distribution.
    """
    calculate_statistics(distribution)
    graph_distribution(distribution)
```

<h5>Uniform Distribution</h5>

```python
distribution = np.random.uniform(low=0, 
                                 high=1,
                                 size=number_of_samples)
summarize(distribution, dist_name='Uniform')
```

<h5>Normal Distribution</h5>

```python
distribution = np.random.normal(size=number_of_samples)
summarize(distribution, dist_name='Normal')
```

<h5>Poisson Distribution</h5>

```python
expected_value=5
distribution = np.random.poisson(expected_value, size=number_of_samples)
summarize(distribution, dist_name='Poisson')
```

<h5>Exponential Distribution</h5>

```python
distribution = np.random.exponential(size=number_of_samples)
summarize(distribution)
```
